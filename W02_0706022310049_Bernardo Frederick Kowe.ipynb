**Handle missing values**

```python
# Menghapus beberapa kolom yang kurang berguna seperti url title dn address karena title dan url tidak perlu digunakan untuk analisa data dan address karena terlalu luas saya akan menggunakan district saja
df = df.drop(columns=['url', 'title', 'address'])

# Propety type hanya rumah jadi tidak berguna di drop saja
df.drop(columns=['property_type'], inplace=True, errors='ignore')

# Hanya menggunakan tahun dibangunnya saja karena nantinya akan diisi dengan median untuk tahun bangunnya jadi sebenernya umur bangunan bisa dihitung dari hari ini ketika seluruh tahun dibangunn nanti diisi dengan median
# Jadi building age tidak di perlukan lagi apalagi melihat data di sheets semua kolom tahun dibangun danumur bangunan selalu null di row yang sama.
df.drop(columns=['building_age'], inplace=True, errors='ignore')

# Kolom numerik yang memiliki nilai null yang kecil di ganti dengan median karena dengan median data lebih tahan terhadap outlier
num_cols_median = ['land_size_m2', 'building_size_m2', 'floors', 'year_built']
for col in num_cols_median:
    df[col].fillna(df[col].median(), inplace=True)

# Kolom kategorikal yang memiliki nilai null yang kecil di ganti dengan mode
cat_cols_mode = ['ads_id', 'certificate', 'bedrooms', 'bathrooms']
for col in cat_cols_mode:
    df[col].fillna(df[col].mode()[0], inplace=True)

# Untuk kolom dengan nilai null yang cukup banyak akan di ganti dengan Unknown karena jika di ganti dengan median atau mode akan menimbulkan kerusakan data karena yang diubah terlalu banyak
df['furnishing'].fillna('Unknown', inplace=True)

# Building orientation dan property condition merupakan data kategorikal dan menurut saya merupakan data yang cukup mempengaruhi harga jual rumah sehingga walaupun nilai nullnya banyak akan saya buat menjadi unknown
df['building_orientation'].fillna('Unknown', inplace=True)
df['property_condition'].fillna('Unknown', inplace=True)
```

```python
# Dilakukan pengecekan outlier kepada beberapa kolom yang cukup penting untuk mengetahui seekstrem apa outlier yang dihasilkan dan nanti akan di cleaning

num_cols = ['price_in_rp', 'bedrooms', 'bathrooms', 'land_size_m2', 'building_size_m2']

plt.figure(figsize=(14,8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[col], color="skyblue")
    plt.title(f"Boxplot {col}")
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

**Handle Outliers**

```python
# Ngedrop harga yang diatas 70M untu menghindari outlier yang kebanyakan dan membuat data tidak baik untuk di pakai.
df = df[df['price_in_rp'] <= 70000000000]

# Ngedrop data jumlah kamar yang jumlahnya di atas 10 karena menurut saya terlalu berlebihan untuk sebuah rumah memiliki kamar sebanyak > 10
df = df[df['bedrooms'] <= 10]
df = df[df['bathrooms'] <= 10]

# Luas tanah saya asumsikan maksimal untuk sebuah rumah adalah 1250m2 dan untuk luas bangunan saya asumsikan untuk sebuah rumah tingkat tiga dengan luas bangunan per lantai 1000m2 sudah cukup untuk sebuah rumah
df = df[(df['land_size_m2'] <= 1250) & (df['building_size_m2'] <= 3000)]

# Rumah dengan umur maksimal 100 tahun dan minimal 0 tahun, selain itu akan di drop karena diatas 100 tahun menurut saya sudah tidka relevan lagi
df = df[(df['year_built'] <= 2022) & (df['year_built'] >= 1922)]

# Untuk garasi dan carports di asumsikan maksimal 10 untuk sebuah rumah
df = df[(df['garages'] <= 10) & (df['carports'] <= 10)]

# Untuk mengatur index dataframe kembali akrena ketika di drop akan lompat lompat indexnya masih sama seperti sebelumnya kalau tidak di reset index
df.reset_index(drop=True, inplace=True)
```

```python
# Untuk mengetahui hasil cleaning untuk outlier apakah berhasil atau masih perlu diubah

num_cols = ['price_in_rp', 'bedrooms', 'bathrooms', 'land_size_m2', 'building_size_m2', 'year_built']

plt.figure(figsize=(14,8))
for i, col in enumerate(num_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[col], color="skyblue")
    plt.title(f"Boxplot {col}")
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

**Feature Engineering**

```python
# Untuk menghitung harga per meter persegi dari setiap properti, sehingga bisa membandingkan nilai relatif rumah dengan ukuran tanah berbeda
df['price_per_m2'] = np.where(
    df['land_size_m2'] > 0,
    df['price_in_rp'] / df['land_size_m2'],
    np.nan
)

# Rumah dibagi menjadi kategori harga (Low-end, Mid-range, High-end, Luxury) untuk memudahkan segmentasi pasar dan analisis distribusi harga.
def price_category(x):
    if x < 1e9: return 'Low-end'
    elif x < 3e9: return 'Mid-range'
    elif x < 1e10: return 'High-end'
    else: return 'Luxury'
df['price_category'] = df['price_in_rp'].apply(price_category)

# Rumah dikategorikan berdasarkan luas tanah (Small, Medium, Large, Extra Large) agar lebih mudah melihat pola ukuran properti.
def size_category(x):
    if pd.isna(x): return 'Unknown'
    elif x < 100: return 'Small'
    elif x < 200: return 'Medium'
    elif x < 500: return 'Large'
    else: return 'Extra Large'
df['size_category'] = df['land_size_m2'].apply(size_category)

# Rumah dikategorikan berdasarkan luas bangunan (Very Small, Small, Medium, Large, Extra Large) agar lebih mudah melihat pola ukuran properti.
def building_size_category(x):
    if pd.isna(x):
        return 'Unknown'
    elif x < 50: return 'Very Small'   # kos/rumah kecil
    elif x < 100: return 'Small'        # tipe 36/45/60
    elif x < 200: return 'Medium'       # rumah menengah
    elif x < 500: return 'Large'        # rumah besar / ruko
    else: return 'Extra Large'  # villa / gedung besar
df['building_size_category'] = df['building_size_m2'].apply(building_size_category)

# Menghitung usia rumah dari tahun dibangun dan membaginya ke kategori seperti Brand New, New, Relatively New, Moderate Age, Old, Very Old, agar analisis usia properti lebih mudah tanpa harus memiliki kolom building_age.
def age_category(year):
    if pd.isna(year) or year <= 0:
        return 'Unknown'
    age = 2022 - year
    if age <= 2:
        return 'Brand New'
    elif age <= 5:
        return 'New'
    elif age <= 10:
        return 'Relatively New'
    elif age <= 25:
        return 'Moderate Age'
    elif age <= 50:
        return 'Old'
    else:
        return 'Very Old'

df['age_category'] = df['year_built'].apply(age_category)

# Jumlah kamar tidur dibagi menjadi Small, Medium, Large untuk memudahkan analisis hubungan jumlah kamar dengan harga atau ukuran rumah.
def bedroom_category(x):
    if pd.isna(x): return 'Unknown'
    elif x <= 2: return 'Small'
    elif x <= 4: return 'Medium'
    else: return 'Large'
df['bedroom_category'] = df['bedrooms'].apply(bedroom_category)
```
